```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel(r'final_try1.xlsx')
```

```python
data.head()
```

```python
data.info()
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
data['Airline'].value_counts()
```

```python
Airline = data[['Airline']]

Airline = pd.get_dummies(Airline, drop_first = True)
Airline.head()
```

```python
data['Origin'].value_counts()
```

```python
Origin = data[['Origin']]

Origin = pd.get_dummies(Origin, drop_first = True)

Origin.head()
```

```python
data['Destination'].value_counts()
```

```python
Destination = data[['Destination']]

Destination = pd.get_dummies(Destination, drop_first = True)

Destination.head()
```

```python
# Concatenate dataframe --> data + Airline + Origin + Destination

new_data = pd.concat([data, Airline, Origin, Destination], axis = 1)
```

```python
new_data.shape
```

```python
new_data.head()
```

```python
new_data.drop(["Airline", "Origin", "Destination"], axis = 1, inplace = True)
```
